# Import da 2 base Corretoras

```python
url_corretoras='https://brasilapi.com.br/api/cvm/corretoras/v1' 
response_corretoras=requests.get(url_corretoras)
response_corretoras

if response_corretoras.status_code==200:
    data_corretoras=response_corretoras.json()
else:
    print ("Erro ao acessar a API")

df_corretoras = pd.DataFrame(data_corretoras)
df_corretoras.head(200)
```

# Import da 3 base Cidades

```python
url_cidades='https://brasilapi.com.br/api/cptec/v1/cidade' 
response_cidades=requests.get(url_cidades)
response_cidades

if response_cidades.status_code==200:
    data_cidades=response_cidades.json()
else:
    print ("Erro ao acessar a API")

df_cidades = pd.DataFrame(data_cidades)
df_cidades.head(200)
```

# Import da 4 base Previsao

```python
clima_data= []

for city_id in df_cidades['id']:
    url_clima =f'https://brasilapi.com.br/api/cptec/v1/clima/previsao/'+ str(city_id)
    response_clima=requests.get(url_clima)
    data_clima=response_clima.json()

    clima_data.append(data_clima)

df_clima = pd.DataFrame(clima_data)
df_clima.head(10)
```

# Cruzamento das duas bases:  Cidades e Previsão (Left)

```python
df_merge= df_cidades.merge(df_clima,left_on='nome',right_on='cidade',how='left')
df_merge.head(10)
```
